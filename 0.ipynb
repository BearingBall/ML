```python
import ZeroConstructions as zc
import matplotlib.pyplot as plt
import numpy as np
%load_ext autoreload
%autoreload 2
```

```python
classifiers = zc.getBinaryClassifiers()
classifiers = zc.CalClassifiers(classifiers)
dataset = zc.getBinaryDataset()
zc.drawBinaryDataset(dataset)
X_train, X_test, y_train, y_test = zc.splitBinaryDataset(dataset)
```

```python
zc.fitBinaryClassifiers(classifiers, X_train, y_train)
```

```python
obj = zc.predictionTable(classifiers[0], X_test)
    
ind = np.argsort(obj)
obj = obj[ind]
y = y_test[ind]

prev = -1.0
for i in range(len(obj)):
    if (prev != obj[i]):
        print(obj[i], y[i])
        prev = obj[i]
```

```python
#zc.drawRocTable(np.array([0.1,0.4,0.6,0.8]), np.array([0,0,1,1]))
#zc.drawRocTable(np.array([1, 1, 1, 0, 0]), np.array([1, 1, 1, 0, 0]))
zc.drawRocTable(obj, y)
```

```python
#zc.drawPRTable(np.array([0.1,0.4,0.6,0.8]), np.array([0,0,1,1]))
#zc.drawPRTable(np.array([1, 1, 1, 0, 0]), np.array([1, 1, 1, 0, 0]))
zc.drawPRTable(obj, y)
```

```python
pred, obj = zc.predictCoinFlipper(y_test)
prediction = [obj[i] if pred[i] == 1 else 1-obj[i] for i in range(len(pred))]

trueV = [1 if pred[i] == y_test[i] else 0 for i in range(len(pred))]
print(np.sum(trueV)/len(pred))

zc.drawRocTable(np.array(prediction), y_test)
```

```python
zc.drawPRTable(np.array(prediction), y_test)
```
